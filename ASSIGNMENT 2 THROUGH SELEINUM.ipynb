# Q1 # Data analyst web scrapping

```python
driver=webdriver.Chrome(r"C:\Users\Mauna\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.XPATH,'//div[@class="qsbSubmit"]')
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scrapping the Job title
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
#Scrapping the Job locations
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for j in location_tags[0:10]:
    location=j.text
    job_location.append(location)
#Scrapping the Company 
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for k in company_tags[0:10]:
    company=k.text
    company_name.append(company)
#Scrapping the Experience 
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for l in experience_tags[0:10]:
    exp=l.text
    experience_required.append(exp)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import pandas as pd
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company_Name':company_name,'Experience':experience_required})
df
```

# Q4 SUNGLASSES SCRAPPING IN FLIP KART


```python
driver=webdriver.Chrome(r"C:\Users\Mauna\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
product=driver.find_element(By.CLASS_NAME,"_3704LK")
product.send_keys('sunglasses')
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
name=[]
description=[]
price=[]
discount=[]
```

```python
#for page selection
start=0
end=2
for page in range(start,end):
    names=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    descriptions=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    discounts=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in names:
        name.append(i.text)
    for j in descriptions:
        description.append(j.text)
    for k in prices:
        price.append(k.text)
    for l in discounts:
        discount.append(l.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
tnames=name[0:100]
tdes=description[0:100]
tprice=price[0:100]
tdisc=discount[0:100]
```

```python
import pandas as pd
df4=pd.DataFrame({'Brand Name':tnames,'Information':tdes,'Cost':tprice,'Offers':tdisc})
df4       
```
